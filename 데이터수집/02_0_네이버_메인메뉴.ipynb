### 네이버 메뉴 전체 크롤링

```python
lis = bs_obj.findAll('li',{'class':'nav_item'})
len(lis)
```

```python
for li in lis :
    a_tag = li.find('a')
    print(a_tag.text, " : ", a_tag['href'])
```

```python
# 네이버 메뉴 정보 저장
menu=[]
url=[]
```

```python
import pandas as pd
df = pd.DataFrame({'메뉴':menu, 'url':url})
df
```

###  네이버 뉴스 크롤링
#### 네이버 뉴스는 네이버 정책에 따라 모든 언론사들의 뉴스가 랜덤하게 배치됨
- 단, 로그인 후 구독을 추가하면 구독한 언론사들의 뉴스가 나옴

- 2022년7월26일 기준

```python
df = pd.read_csv('./crawl_data/naver_menu.csv')
```

```python
df.head()
```

```python
df.head(1)
url = df[df.메뉴 == '뉴스'].url.values[0] # 문자열 네이버 뉴스 url이 추출
```

```python
print(url)
# 네이버 뉴스 홈페이지 요청
html = urlopen(url)
```

```python
# bs4 객체 생성
bs_obj = bs4.BeautifulSoup(html_text, 'html.parser')
```

```python
print(bs_obj.prettify())
```

```python
# 네이버 뉴스 홈페이지의 왼쪽 카드 뉴스의 타이틀 수집
news_title = bs_obj.findAll('div',{'class':'cjs_t'})
```

```python
len(news_title)
```

```python
for title in news_title :
    print(title.text)
```

```python
# 네이버 뉴스 홈페이지의 왼쪽 카드 뉴스의 세부내용 수집
news_desc = bs_obj.findAll('p',{'class':'cjs_d'})
```

```python
len(news_desc)
```

```python
for desc in news_desc :
    print(desc.text)
```

## 네이버 뉴스 섹션 메뉴와 섹션별 url 추출

```python
from urllib.request import urlopen
import bs4
import pandas as pd
```

```python
url = 'https://news.naver.com'
html = urlopen(url)
bs_obj = bs4.BeautifulSoup(html, 'html.parser')
```

```python
# 네이버 뉴스 섹션메뉴 태그 확인(개발자도구)
# body > section > header > div.Nlnb._float_lnb > div > div > div.Nlnb_left._lnb_scroll > div > div > ul > li:nth-child(3)
# select가 너무 길어서 유용하지 않음
# 적접 확인한 태그의 클래스와 속성 사용
# ul .Nlnb_menu_list
uls = bs_obj.findAll('ul',{'class':'Nlnb_menu_list'})
len(uls)
ul = bs_obj.find('ul',{'class':'Nlnb_menu_list'})
ul
```

```python
lis = ul.findAll('li')
len(lis)
```

```python
for li in lis :
    a_tag = li.find('a')
    print(a_tag.text, ":", a_tag['href'])
```

```python
# li 태그를 이용해서 추출
lis = bs_obj.findAll("li",{"class":"Nlist_item"})
len(lis)
```

```python
# 수집데이터 df 구성
section=[]
link =[]
```

```python
for li in lis :
    a_tag = li.find('a')
    section.append(a_tag.text)
    link.append(a_tag['href'])
```

```python
col_dict ={'section':section, 'link':link}
```

```python
news_section_df = pd.DataFrame(col_dict)
```

```python
news_section_df
```

```python
news_section_df.to_csv('./crawl_data/naver_news_section.csv')
```
